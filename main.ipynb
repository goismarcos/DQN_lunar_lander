```python
import gym
import matplotlib.pyplot as plt
import numpy as np
from agent import Agent
from train import Train
import warnings
warnings.filterwarnings('ignore')
```

## Instanciando o ambiente e descobrindo a quantidade de estados presente e números de ações possíveis

```python
env = gym.make('LunarLander-v2')
env.seed(0)

print('Quantidade de estados: ', env.observation_space.shape[0])
print('Número de ações: ', env.action_space.n)
```

## Obtendo o agente, fazendo seu treinamento e salvando os scores

```python
agent = Agent(state_size=8, action_size=4, seed=0, ddqn=False)
train = Train(env, agent)
scores, avg_scores = train.train()
```

## Plotando o gráfico da quantidade de episódios necessários no treinamento para obter o score 200  

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.plot(np.arange(len(avg_scores)), avg_scores, label=' avg',color = 'blue')
plt.ylabel('Score')
plt.xlabel('Episódio #')
plt.show()
```

## Mostrando um vídeo do Lander já treinado fazendo pouso

```python
train.show_video_of_model(agent, 'LunarLander-v2')
train.show_video('LunarLander-v2')
```
